```python
x = [4, 8, 6, 16,22, 27, 36, 49, 53, 70, 82, 95, 109]
y = [25, 50, 75, 100, 125, 150, 175, 200, 225, 250, 275, 300, 325]
```

```python
def rata(data):
    total = 0
    for i in data:
        total += i
    
    hasil = total / len(data)
    return hasil
    #return sum(data) / len(data)

def cariB(x, y, rataX, rataY):
    # Atas
    n = len(x)
    totalatas = 0
    for i in range(n):
        totalatas += (x[i] - rataX) * (y[i] - rataY)
    # Bawah
    totalbawah = 0
    for i in range(n):
        totalbawah += (x[i] - rataX) ** 2

    # Atas / bawah
    return totalatas / totalbawah

def cariA(rataY, rataX, B):
    return rataY - (B * rataX)

def cariY(A, B, x):
    return A + (B * x)
```

```python
rataX = rata(x)
rataY = rata(y)

B = cariB(x, y, rataX, rataY)
A = cariA(rataY, rataX, B)

print(f"Y = {A} + {B}X")
print(f"Y = {A} + {B}*6 = {cariY(A, B, 6)}")
```
