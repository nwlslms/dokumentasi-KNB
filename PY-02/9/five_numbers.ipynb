```python
# Data set
data = [6, 3, 1, 1, 4, 5, 10, 9, 8, 7]
```

```python
# Fungsi 5 Number Summary
def summary(data):
    print("Data belom terurut: ", end='')
    for i in data:
        print(i, end=' ')
    print()
    data.sort()
    n = len(data)
    print("Data terurut: ", end='')
    for i in data:
        print(i, end=' ')
    print()
    # Min
    min = data[0]

    # Max
    max = data[-1]
    
    # Median (Q2)
    if n % 2 == 0:
        # 10 / 2 = 5
        # Data ke - 5 dan ke - 6
        # Tapi karena akses index, maka dikurang 1
        q2 = (data[n//2 - 1] + data[n//2]) / 2
    else:
        q2 = data[n//2]

    # Q1
    data_kiri = data[:n//2]
    if len(data_kiri) % 2 == 0:
        q1 = (data_kiri[len(data_kiri) // 2 - 1] + data_kiri[len(data_kiri) //2 ]) / 2
    else:
        q1 = data_kiri[len(data_kiri) // 2]

    # Q3
    if n % 2 == 0:
        data_kanan = data[n//2:]
    else:
        data_kanan = data[n//2+1:]
    
    if len(data_kanan) % 2 == 0:
        q3 = (data_kanan[len(data_kanan) // 2 - 1] + data_kanan[len(data_kanan) // 2]) / 2
    else:
        q3 = data_kanan[len(data_kanan) // 2]
    print(f"Min: {min}\nQ1: {q1}\nQ2: {q2}\nQ3: {q3}\nMax: {max}")
```

```python
summary(data)
```
