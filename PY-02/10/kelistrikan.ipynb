```python
# Deklarasi variabel
v = -1
i = -1
r = -1
```

```python
# Deklarasi Fungsi
def ohm(v, i, r):
    if v == -1:
        if i != -1 and r != -1:
            v = i * r
            return f"v = {i} * {r}  = {v}"
        else:
            print("Informasi kurang!")
    elif i == -1:
        if v != -1 and r != -1:
            i = v / r
            return f"i = {v} / {r} = {i}"
        else:
            print("Informasi kurang!")
    elif r == -1:
        if v != -1 and i != -1:
            r = v / i
            return f"r = {v} / {i} = {r}"
        else:
            print("Informasi kurang!")
    else:
        print("Ketemu semua")

def daya(v, i, r):
    if v != -1 and i != -1:
        p = v * i
        return f"p = {v} * {i} = {p}"
    elif i != -1 and r != -1:
        p = (i**2) * r
        return f"p = {i}^2 * {r} = {p}"
    elif v != -1 and r != -1:
        p = (v**2) / r
        return f"p = {v}^2 / r"
    
def seri(resistensi):
    #total = sum(resistensi)
    total = 0
    for i in resistensi:
        total += i
    return f"R = {total}"

def parerel(resistensi):
    total = 0
    for i in resistensi:
        total += (1/i)
    total = 1 / total
    return f"R = {total}"
```

```python
# Eksekusi
print(ohm(24, -1, 12))
print(daya(220, 0.5, -1))
print(seri([5, 10, 15]))
print(parerel([12, 6]))
print(ohm(220, -1, 44))
print(daya(220, 5, -1))
```
