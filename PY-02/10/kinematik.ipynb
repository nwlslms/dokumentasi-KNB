```python
# Deklarasi variabel
import math as m
s = -1
v = -1
t = -1
a = 4
vt = 0
v0 = 12
```

```python
# Deklarasi Fungsi
def glb(s, v, t):
    if s == -1:
        if v != -1 and t != -1:
            s = v * t
            return f"s = {v} * {t} = {s}"
        else:
            return f"Informasi kurang!"
    elif v == -1:
        if s != -1 and t != -1:
            v = s / t
            return f"v = {s} / {t} = {v}"
        else:
            return f"Informasi kurang!"
    elif t == -1:
        if s != -1 and v != -1:
            t = s / v
            return f"t = {s} / {v} = {t}"
        else:
            return f"Informasi kurang!"
    else:
        return f"Semua sudah diketahui!"
    
def glbb(s, t, a, v0, vt):
    # Nyari jarak
    if s == -1:
        if v0 != -1 and t != -1 and a != -1:
            s = (v0 * t) + ((a*(t**2))/2)
            return f"s = {v0} * {t} + ({a}*{t}^2) / 2 = {s}"
        elif vt != -1 and a != -1 and v0 != -1:
            s = (vt**2 - v0**2) / (2*a)
            return f"s = ({vt}^2 - {v0}^2) / 2*{a} = {s}"
    # Nyari waktu
    elif t == -1:
        if vt != -1 and v0 != -1 and a != -1:
            t = (vt - v0) / a
            return f"t = ({vt} - {v0}) / {a} = {t}"
    # Nyari percepatan
    elif a == -1:
        if vt != -1 and v0 != -1 and t != -1:
            a = (vt - v0) / t
            return f"a = ({vt} - {v0} / {t}) = {a}"
        elif vt != -1 and v0 != -1 and s != -1:
            a = ((vt**2) - (v0**2)) / (2*s)
            return f"a = ({vt}**2 - {v0}**2) / (2*{s}) = {a}"

    # Nyari kecepatan awal
    elif v0 == -1:
        if vt != -1 and a != -1 and t != -1:
            v0 = vt - (a * t)
            return f"v0 = {vt} - ({a} * {t}) = {v0}"
        elif vt != -1 and a != -1 and s != -1:
            v0 = m.sqrt((vt**2) - (2*a*s))
            return f"v0 = sqrt({vt}^2 - (2*{a}*{s}))"
    # Nyari kecepatan akhir
    elif vt == -1:
        if v0 != -1 and a != -1 and t != -1:
            vt = v0 + (a * t)
            return f"vt = {v0} + ({a}*{t}) = {vt}"
        elif v0 != -1 and a != -1 and s != -1:
            vt = m.sqrt((v0**2) + (2*a*s))
            return f"vt = sqrt({v0}^2 + 2*{a}*{s} = {vt})"
    else:
        return "Kombinasi tidak lengkap dan tidak sesuai"
```

```python
# Eksekusi
print(glb(-1, 20, 15))
print(glb(200, 8, -1))
print(glbb(-1, 10, 2, 0, -1))
print(glbb(-1, 4, 3, 5, -1))
print(glbb(50, -1, 2, 10, -1))
```
