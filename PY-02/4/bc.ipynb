```python
#break
while True:
    a = int(input("a: "))
    if a == 0:
        break
    b = int(input("b: "))
    if b == 0:
        break
    print(f"{a} + {b} = {a + b}")
```

```python
#Continue
i = 0
while True:
    if i % 2 == 0:
        i += 1
        continue
    print(i)
    i+=1
    if i == 100:
        break
```
